```python
# input-list

str_list = ['a', 'b', 'c']
```

```python
# str-parallel-processing

new_lines = []
for line in str_list:
    line += '_processed'
    print(line)
    new_lines.append(line)
```

```python
# assert-new-lines

assert new_lines == ['a_processed', 'b_processed', 'c_processed']
```
