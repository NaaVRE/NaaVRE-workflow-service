```python
import os

conf_data_folder = os.path.join('/tmp', 'data')
conf_user_folder = '/home/jovyan/Cloud Storage/naa-vre-user-data/'
param_with_dash = '-param'
secret_password = 'secret'
```

```python
# produce-types
int_val = 1
int_type = 'int'

float_val = 1.0
float_type = 'float'

string_val ='hello.txt'
string_type = 'str'

list_int_val = [1,2,3]
list_int_type = 'list'

list_double_val = [1.0,2.0,3.0]
list_double_type = 'list'

list_str_val = ["1","2","3"]
list_str_type = 'list'
```

```python
# write-to-file


os.makedirs(conf_data_folder, exist_ok=True)

file_path = os.path.join(conf_user_folder, string_val)

with open(file_path, "w", encoding="utf-8") as f:
    f.write("Created using write mode.\n")
    f.write("Second line.\n")
    f.write("Third line.\n")
```

```python
# process-file

lines = []
with open(file_path) as file:
    while line := file.readline():
        lines.append(line.rstrip())
```

```python
# str-parallel-processing

new_lines = []
for line in lines:
    line += '_processed'
    print(line)
    new_lines.append(line)
```

```python
# int-parallel-processing

new_list_int_val = []
for elem in list_int_val:
    new_list_int_val.append(elem+1)
```

```python
# double-parallel-processing

new_list_double_val = []
for double_val in list_double_val:
    new_list_double_val.append(double_val+1.1)
```

```python
# assert


assert ['Created using write mode._processed', 'Second line._processed', 'Third line._processed'] == new_lines


if not secret_password:
    raise ValueError('secret_password is empty.') 
if not conf_data_folder:
    raise ValueError('conf_data_folder is empty.') 
if not conf_user_folder:
    raise ValueError('conf_user_folder is empty.') 
if not param_with_dash:
    raise ValueError('param_with_dash is empty.')

assert secret_password == 'secret'
assert conf_data_folder == os.path.join('/tmp', 'data')
assert param_with_dash == '-param'


def check_type(value, expected_types: str):
    type_map = {
        "int": int,
        "float": float,
        "str": str,
        "bool": bool,
        "list": list,
        "tuple": tuple,
        "dict": dict,
        "set": set,
    }

    expected_type_names = [t.strip() for t in expected_types.split(",")]

    try:
        expected_type_objects = tuple(type_map[name] for name in expected_type_names)
    except KeyError as e:
        raise ValueError(f"Unsupported type name: {e.args[0]}")

    if not isinstance(value, expected_type_objects):
        raise TypeError(
            f"Expected type(s) {expected_types}, "
            f"but got {type(value).__name__}"
        )

    return True


check_type(int_val,int_type)
check_type(float_val,float_type)
check_type(string_val,string_type)
check_type(list_int_val,list_int_type)
check_type(list_double_val,list_double_type)
check_type(list_str_val,list_str_type)


assert os.path.exists(file_path)

content = ''
with open(file_path, "r") as file:
    content += file.read()

assert content == 'Created using write mode.\nSecond line.\nThird line.\n'
assert new_list_int_val == [2, 3, 4]
assert new_list_double_val == [2.1, 3.1, 4.1]
```
