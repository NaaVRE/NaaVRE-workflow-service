```python
# Create a JSON
# Import built-in json module
import json

# A dictionary
data = {"task": "Write Python", "number": 1, "skills": ["python", "sql"]}

# serialize with json.dumps method
json_str = json.dumps(data)

print(json_str)
```

```python
# Use the JSON
# Deserialize with json.loads method
data = json.loads(json_str)

print(data)

original_data = {"task": "Write Python", "number": 1, "skills": ["python", "sql"]}
assert data == original_data
```
